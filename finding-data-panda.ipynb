# Loading Local Data in Google Colab

Since you are running this in **Google Colab**, the code runs on Google's cloud servers, not your local computer. This means the notebook cannot see `anime.csv` on your computer unless you **upload it** first.

## Option 1: Drag and Drop
1. Click the file folder icon on the left sidebar in Colab.
2. Drag `anime.csv` from your computer into that sidebar.

## Option 2: Upload via Code
Run the cell below to verify if the file is present, or upload it if it's missing.

```python
import pandas as pd
import os

file_name = 'anime.csv'

# Check if running in Colab and file is missing
try:
    from google.colab import files
    if not os.path.exists(file_name):
        print(f"{file_name} not found locally.")
        try:
            print("Attempting to use the upload widget...")
            print("(If this hangs or fails in VS Code, stop it and upload manually via the file sidebar)")
            uploaded = files.upload()
            if file_name not in uploaded.keys() and len(uploaded) > 0:
                print(f"Note: You uploaded {list(uploaded.keys())[0]}, but we were looking for {file_name}.")
        except Exception:
            print("\n[NOTICE] Automatic upload failed or was interrupted (common in VS Code).")
            print(f"Please manually drag and drop '{file_name}' into the file explorer sidebar on the left.")
    else:
        print(f"{file_name} is found!")
except ImportError:
    print("Not running in Google Colab environment. Ensure the file is in the same directory.")

# Now load the data
if os.path.exists(file_name):
    try:
        df = pd.read_csv(file_name)
        print("\nSuccess! Data loaded.")
        display(df.head())
    except Exception as e:
        print(f"Error reading file: {e}")
else:
    print(f"\nError: {file_name} not found. Please upload it using the File sidebar.")
```
