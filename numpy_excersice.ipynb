```python
import numpy as np 
s = np.array([
    [5,3,4,6,7,8,9,1,2],
    [6,7,2,1,9,5,3,4,8],
    [1,9,8,3,4,2,5,6,7],

    [8,5,9,7,6,1,4,2,3],
    [4,2,6,8,5,3,7,9,1],
    [7,1,3,9,2,4,8,5,6],

    [9,6,1,5,3,7,2,8,4],
    [2,8,7,4,1,9,6,3,5],
    [3,4,5,2,8,6,1,7,9]
])
```

```python
b = np.sum(s,axis=1)
for i in b:
    if i != 45:
        print("Sudoku is not valid")
        break
else:
    print("Sudoku row is valid")
```

```python
c = np.sum(s,axis=0)
for i in c:
    if i != 45:
        print("Sudoku is not valid")
        break
else:
    print("Sudoku col is valid")
```

```python
data = np.array([
    [18,85,78],
    [19,92,98],
    [17,76,95],
    [18,65,70],
    [20,90,85]
])
```

```python
data.shape
```

```python
ages = data[0::,0]
average_age = ages.mean()
print(average_age)
```

```python
math_marks = data[0::,1]
math_marks
```

```python
science_marks = data[0::,2]
highest_science_marks = science_marks.max()
print(highest_science_marks)
```

```python
math_marks_idx = np.where(math_marks > 90)
highest_scored_student = data[math_marks_idx]
print(highest_scored_student)
```
