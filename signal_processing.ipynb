The Moving Average

```python
import numpy as np
k = int(input("Enter number of slices you want"))
n = int(input("Enter length of array ?"))
array_1 = np.arange(n)
window_sum = np.sum([array_1[i:i+n-k+1] for i in range(k)],axis=0)
moving_avg = window_sum / k
print(moving_avg[:10])
```

The Nearest Neighbor

```python
landmarks = np.random.randn(10,3)
robot_pos = np.random.randn(3)

the_nearest_neighbour = landmarks - robot_pos
squared_difference = np.square(the_nearest_neighbour)
distances_sq = np.sum(squared_difference, axis=1)
closest_index = np.argmin(distances_sq)

print(f"The closest landmark is at index: {closest_index}")
print(f"The distance is: {np.sqrt(distances_sq[closest_index])}")
```

Image Masking

```python
img = np.random.randint(0, 255, (256, 256, 3))
red_channel = img[:, :, 0]
mask = red_channel > 200
img[mask] = [0,0,225]
print(img)
```
